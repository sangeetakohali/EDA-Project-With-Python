## Data Cleaning

```python
netflix_df.columns
```

```python
netflix_df['Netflix Release Date'] = pd.to_datetime(netflix_df['Netflix Release Date'])
```

```python
netflix_df.dtypes
```

```python
netflix_data = netflix_df.drop(columns=['Netflix Link','IMDb Link','Image','Metacritic Score','Runtime','TMDb Trailer'
                                        ,'Production House','TMDb Trailer'],axis=1)
```

```python
netflix_data.head()
```

## Finding Missing Value

```python
netflix_data.isnull().sum()
```

## Handle Missing Value

```python
# Using pandas fillna(ffill method)
netflix_newdata = netflix_data.fillna(method='ffill')
```

```python
netflix_newdata.isnull().sum()
```

```python
netflix_newdata['Trailer Site'] = netflix_newdata['Trailer Site'].fillna(netflix_newdata['Trailer Site'].mode()[0])
```

```python
netflix_newdata.isnull().sum()
```

## Bivariate Analysis

```python
# Yearly No Of Movie Release
total_release_movie = netflix_newdata[(netflix_newdata.Categeries=='Movie')][['Categeries','Year']]
total_release_movie = total_release_movie.groupby('Year').count().reset_index().sort_values(by='Categeries',ascending=False)
```

```python
total_release_movie
```

```python
plt.figure(figsize=(10,5))
sns.set_style('darkgrid')
sns.barplot(x='Year',y='Categeries',data=total_release_movie,palette='coolwarm')
plt.title('Total Movie Release of Each Year',size=20)
plt.xticks(rotation=70)
plt.xlabel('Year')
plt.ylabel('Total No Of Movie')
```

```python
Series_df = netflix_newdata[netflix_newdata['Categeries'].isin(['Series'])]
```

```python
top_awarded = Series_df.groupby('Title')['Awards Received'].aggregate(max).sort_values(ascending=False).reset_index()
```

```python
# Series Received Awards
plt.figure(figsize=(15,5))
plt.title('Top 10 Series Received Awards',size=25)
sns.barplot(x='Title',y='Awards Received',data=top_awarded.iloc[:10], palette = 'nipy_spectral',edgecolor='black')
plt.xticks(rotation=60)
plt.xlabel('Series Title')
plt.ylabel('No of Awards')
```

```python
movie_df = netflix_newdata[netflix_newdata['Categeries'].isin(['Movie'])]
```

```python
top_movie = movie_df.groupby('Title')['Awards Received'].aggregate(max).sort_values(ascending=False).reset_index()
```

```python
# Movie Received Maximum Awards
plt.figure(figsize=(15,5))
sns.barplot(x='Title',y='Awards Received',data = top_movie.iloc[:10],palette='cubehelix',edgecolor='black')
plt.title('Top 10 Movies Received Awards',size=25)
plt.xticks(rotation=45)
plt.xlabel('Movies Title')
plt.ylabel('No of Awards')
```

## Remove outlires

```python
cols = ['IMDb Votes','Awards Received','IMDb Score']
Q1 = netflix_newdata[cols].quantile(0.25)
Q3 = netflix_newdata[cols].quantile(0.75)
IQR = Q3-Q1
final_netflix_newdata = netflix_newdata[~((netflix_newdata[cols]<( Q1-1.5*IQR )) | (netflix_newdata[cols]> ( Q3+1.5*IQR ))).any(axis=1)]
```
